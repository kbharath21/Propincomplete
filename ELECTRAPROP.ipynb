```python
from sklearn.model_selection import train_test_split

# First, split the data into a smaller test set (100 samples)
train_df, temp_df = train_test_split(df, test_size=0.06, random_state=42)  # 90% for training, 10% for temp (validation + test)
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)  # Split the temp set into 50% validation and 50% test

print(f"Training set size: {len(train_df)}")
print(f"Validation set size: {len(valid_df)}")
print(f"Test set size: {len(test_df)}")
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
# Tokenize the data using the preprocessed text
train_inputs = torch.tensor(train_df['tokens'].tolist())
train_labels = torch.tensor(train_df['label_bias'].tolist())

valid_inputs = torch.tensor(valid_df['tokens'].tolist())
valid_labels = torch.tensor(valid_df['label_bias'].tolist())

test_inputs = torch.tensor(test_df['tokens'].tolist())
test_labels = torch.tensor(test_df['label_bias'].tolist())
```

```python
train_data = TensorDataset(train_inputs, train_labels)
valid_data = TensorDataset(valid_inputs, valid_labels)
test_data = TensorDataset(test_inputs, test_labels)

train_loader = DataLoader(train_data, batch_size=8, shuffle=True)
valid_loader = DataLoader(valid_data, batch_size=8)
test_loader = DataLoader(test_data, batch_size=8)
```

```python
from transformers import ElectraForSequenceClassification, ElectraTokenizer
model = ElectraForSequenceClassification.from_pretrained('google/electra-small-discriminator', num_labels=2)
tokenizer = ElectraTokenizer.from_pretrained('google/electra-small-discriminator')
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
import torch
from torch.optim import AdamW
from tqdm import tqdm
import torch.nn.functional as F

# Define optimizer
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)

# Training loop
epochs = 4  # Running for 4 epochs
for epoch in range(epochs):
    model.train()
    total_loss = 0
    correct_predictions = 0  # Track correct predictions
    total_samples = 0
    batch_count = len(train_loader)  # Total number of batches
    progress_bar = tqdm(enumerate(train_loader), total=batch_count, desc=f"Epoch {epoch+1}")

    for batch_idx, batch in progress_bar:
        batch_inputs, batch_labels = batch
        batch_inputs = batch_inputs.to(device)
        batch_labels = batch_labels.to(device)

        optimizer.zero_grad()
        outputs = model(batch_inputs, labels=batch_labels)
        loss = outputs.loss
        logits = outputs.logits  # Get model predictions

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        # Calculate predictions & accuracy
        predictions = torch.argmax(logits, dim=1)  # Get class with highest probability
        correct_predictions += (predictions == batch_labels).sum().item()
        total_samples += batch_labels.size(0)

        # Update progress bar with batch loss
        progress_bar.set_postfix({"Batch Loss": f"{loss.item():.4f}"})

    # Compute average loss and accuracy for the epoch
    avg_loss = total_loss / batch_count
    epoch_accuracy = (correct_predictions / total_samples) * 100  # Convert to percentage

    print(f"\nEpoch {epoch+1} completed. Average Loss: {avg_loss:.4f} | Accuracy: {epoch_accuracy:.2f}%\n")
```

```python
import torch

model.eval()  # Switch to evaluation mode

valid_loss = 0
correct = 0
total = 0

for batch in valid_loader:
    batch_inputs, batch_labels = batch
    batch_inputs = batch_inputs.to(device)
    batch_labels = batch_labels.to(device)

    with torch.no_grad():
        outputs = model(batch_inputs, labels=batch_labels)
        loss = outputs.loss
        logits = outputs.logits  # Get raw predictions

    valid_loss += loss.item()

    # Compute accuracy
    preds = torch.argmax(logits, dim=1)  # Get predicted class
    correct += (preds == batch_labels).sum().item()
    total += batch_labels.size(0)

accuracy = correct / total
print(f"Validation Loss: {valid_loss / len(valid_loader):.4f}")
print(f"Validation Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Initialize lists for true labels and predictions
all_predictions = []
all_labels = []

model.eval()  # Set model to evaluation mode

with torch.no_grad():
    for batch in test_loader:
        batch_inputs, batch_labels = batch
        batch_inputs = batch_inputs.to(device)
        batch_labels = batch_labels.to(device)

        outputs = model(batch_inputs, labels=batch_labels)
        logits = outputs.logits  # Get raw model outputs

        # Compute predictions
        predictions = torch.argmax(logits, dim=1)  # Get class with highest probability

        # Store for classification report
        all_predictions.extend(predictions.cpu().numpy())  # Convert to list
        all_labels.extend(batch_labels.cpu().numpy())

# Generate Classification Report
print("\nClassification Report:")
print(classification_report(all_labels, all_predictions, digits=4))

# Compute Confusion Matrix
conf_matrix = confusion_matrix(all_labels, all_predictions)
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
from scipy.stats import ks_2samp

# Separate positive and negative class probabilities
pos_scores = all_predictions[np.array(all_labels) == 1]
neg_scores = all_predictions[np.array(all_labels) == 0]

# Compute cumulative distributions
pos_cdf = np.sort(pos_scores)
neg_cdf = np.sort(neg_scores)

# KS statistic
ks_stat, _ = ks_2samp(pos_scores, neg_scores)

# Plot KS Curve
plt.figure(figsize=(6, 5))
plt.plot(np.linspace(0, 1, len(pos_cdf)), pos_cdf, label="Positive Class", color="blue")
plt.plot(np.linspace(0, 1, len(neg_cdf)), neg_cdf, label="Negative Class", color="red")
plt.axvline(ks_stat, color="gray", linestyle="--", label=f"KS = {ks_stat:.3f}")
plt.xlabel("Cumulative Distribution")
plt.ylabel("Score")
plt.title("Kolmogorov-Smirnov (KS) Plot")
plt.legend()
plt.show()
```

```python
##### IGNORE ######



import torch
from transformers import AutoTokenizer

# Load tokenizer (replace with your model's tokenizer)
tokenizer = AutoTokenizer.from_pretrained("google/electra-base-discriminator")  # Change if using another variant

def predict_statement(statement, model, device):
    """Preprocess, run inference, and decode output for a given statement."""
    model.eval()  # Set to evaluation mode

    # Tokenize input text
    inputs = tokenizer(statement, return_tensors="pt", truncation=True, padding=True, max_length=512)
    inputs = {key: value.to(device) for key, value in inputs.items()}  # Move tensors to correct device

    with torch.no_grad():  # No gradient calculation needed
        outputs = model(**inputs)

    logits = outputs.logits
    probabilities = torch.softmax(logits, dim=1)  # Convert logits to probabilities
    predicted_class = torch.argmax(probabilities, dim=1).item()  # Get predicted label index

    return predicted_class, probabilities.cpu().numpy()  # Return class and probabilities

# Example: Running predictions on test statements
for statement in test_statements:
    predicted_class, probabilities = predict_statement(statement, model, device)
    print(f"Statement: {statement}")
    print(f"Predicted Class: {predicted_class}")
    print(f"Probabilities: {probabilities}")
    print("-------------------------------------------------")
```

 **DeBERTa's tokenizer**

```python


import torch
from transformers import DebertaTokenizerFast, DebertaForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset, Dataset
from sklearn.model_selection import train_test_split
import pandas as pd

# Load dataset
dataset_path = "FINALSG2.xlsx"  # Ensure the file is uploaded
df = pd.read_excel(dataset_path)

# Preprocessing
df = df.dropna(subset=['text', 'label_bias'])
df['label_bias'] = df['label_bias'].map({'Non-biased': 0, 'Biased': 1, 'No agreement': 0})

# Split dataset
train_texts, val_texts, train_labels, val_labels = train_test_split(df['text'], df['label_bias'], test_size=0.2, random_state=42)

# Load DeBERTa tokenizer
tokenizer = DebertaTokenizerFast.from_pretrained("microsoft/deberta-base")

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Convert to Hugging Face Dataset
train_data = Dataset.from_dict({"text": train_texts.tolist(), "label": train_labels.tolist()})
val_data = Dataset.from_dict({"text": val_texts.tolist(), "label": val_labels.tolist()})
train_data = train_data.map(tokenize_function, batched=True)
val_data = val_data.map(tokenize_function, batched=True)

# Load DeBERTa model
model = DebertaForSequenceClassification.from_pretrained("microsoft/deberta-base", num_labels=2)

# Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data,
)

# Train the model
trainer.train()

# Save the trained model
model.save_pretrained("./deberta_model")
tokenizer.save_pretrained("./deberta_tokenizer")
```
